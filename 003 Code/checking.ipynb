```python
from tqdm import tqdm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import os
import re
from torch import nn
import torch

class AutoEncoder(nn.Module):
    def __init__(self):
        super(AutoEncoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(200, 64),
            nn.ReLU(),
            nn.Linear(64, 32)
        )
        self.decoder = nn.Sequential(
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 200)
        )

    def forward(self, x):
        encoded = self.encoder(x)
        return self.decoder(encoded)

name_list = glob("preprocessing_A/*")
for i in range(len(name_list)):
    name_list[i] = name_list[i].split("/")[1]
name_list.sort()

file_list = glob(f"preprocessing_A/{name_list[0]}/*.csv")
cel_regex = re.compile("^cel*")

csv_file = pd.read_csv(f"{file_list[0]}", encoding = 'cp949')
csv_file.drop(columns = "Unnamed: 0", inplace = True)
csv_file[csv_file.columns[0]] = pd.to_datetime(csv_file[csv_file.columns[0]])

start = csv_file[csv_file.columns[0]].min()
end = csv_file[csv_file.columns[0]].max()
new_range = pd.date_range(start, end, freq = "min")
new_dataframe = pd.DataFrame({csv_file.columns[0]:new_range})
csv_file = pd.merge(csv_file, new_dataframe, on = csv_file.columns[0], how = "right")

new = csv_file.loc[:, [csv_file.columns[0], "string_no", "rack_no", "module_no"]]

for i in csv_file.columns:
  if cel_regex.match(i): 
    concat_df = csv_file.loc[:, i]
    new = pd.concat([new, concat_df], axis = 1)
```
